```python
from pathlib import Path
from collections import defaultdict

def count_files_per_folder(root_dir: Path) -> dict[str, int]:
    """Count files in each subfolder under root_dir."""
    counts = defaultdict(int)
    root_dir = Path(root_dir)
    for path in root_dir.rglob("*"):
        if path.is_file():
            # Get the immediate subfolder of root_dir (e.g., train/joy -> joy)
            rel = path.relative_to(root_dir)
            folder = rel.parts[0] if len(rel.parts) > 1 else "(root)"
            counts[folder] += 1
    return dict(counts)

data_dir = Path("data")
for split in ["train", "test"]:
    split_path = data_dir / split
    if split_path.exists():
        counts = count_files_per_folder(split_path)
        print(f"\n{split.upper()}:")
        for folder, n in sorted(counts.items()):
            print(f"  {folder}: {n}")
        print(f"  Total: {sum(counts.values())}")
```
